```python
S = input()
alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
num = []
check = []
check_2 = 0
for i in range(0,26) :
    num.append(-1)
for i in range(len(S)) :
    check_2 = check_2 + 1
    for j in range(0,26) :
        if S[i] == alphabet[j]:
            check.append(S[i])
            check = set(check)
            if len(check) == check_2 : 
                num[j] = i
            else : 
                check.append(i)
```
