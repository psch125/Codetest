# while 문

```python
i = 1
result = 0
while i <= 9 : # i가 9보다 작거나 같을 때 아래 코드를 반복적으로 실행
    result += i
    i += 1
print('1~9까지의 합',result)
i = 1
result = 0
while i<= 9 : # i가 9보다 작거나 같을 때 아래 코드를 반복적으로 실행
    if i % 2 == 1 :
        result += i
    i += 1
print('1~9까지 중 홀수의 합',result)
```

# for 문

```python
result = 0
for i in range(1,10) : # i는 1부터 9까지의 모든 값을 순화
    result += i
print('1~9까지의 합',result)
scores = [90,85,77,65,97]
for i in range(5) :
    if scores[i] >= 80 :
        print(i+1,'번 학생은 합격입니다.')
scores = [90,85,77,65,97]
cheating_list = {2,4}
for i in range(5) :
    if i+i in cheating_list :
        continue
    if scores[i] >= 80 :
        print(i+1,'번 학생은 합격입니다.')
#중첩 for문(반복문)
for i in range(2,10) :
    for j in range(1,10) :
        print(i,'X',j,'=',i*j)
```
