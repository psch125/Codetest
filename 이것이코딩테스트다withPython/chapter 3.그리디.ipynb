```python
N = int(input())
plans = (input().split())
x,y = 1,1
move_types = ['L','R','U','D']
for plan in plans :
    for i in range(len(move_types)) :
        if plan == 'L' :
            if y-1 != 0 :
                y = y-1
        if plan == 'R' :
            if y+1 < N :
                y = y+1
        if plan == 'U' :
            if x-1 != 0 :
                x = x-1
        if plan == 'D' :
            if x+1 < N :
                x = x+1
print(x,y)
                
```
