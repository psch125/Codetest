```python
#실수형
a = 157.93 # 양의 실수
print('양의 실수', a)
a = -1837.2 # 음의 실수
print('음의 실수',a)
#소수부가 0일 때 0을 생략
a = 5.
print('소수부가 0일 때 0을 생략',a)
#정수부가 0일 때 0을 생략
a = -.7
print('정수부가 0일 때 0을 생략',a)
a = 1e9 # 10억의 지수 표현 방식
print('10억의 지수 표현 방식',a)
a = 75.25e1 # 752.5
print(a)
a= 3954e-3 # 3.954
print(a)

a = 0.3+0.6
print(a)

if a == 0.9:
    print(True)
else :
    print(False)

# round 함수 사용    
a = 0.3 + 0.6
print('a값을 소수점 다섯째 자리에서 반올림',round(a,4))

if round(a,4) == 0.9:
    print(True)
else :
    print(False)
    
# 수 자료형의 연산
a = 7
b = 3
print('나누기',a/b) # 나누기
print('나머지',a%b) # 나머지
print('몫',a // b) # 몫
a = 5
b = 3
print('거듭제곱',a**b) #거듭제곱
```

# 리스트 자료형

```python
#리스트 만들기
a = [1,2,3,4,5,6,7,8,9]
print(a)
print('인덱스 4, 즉 다섯 번째 원소에 접근',a[4]) #인덱스 4, 즉 다섯 번째 원소에 접근
a = list() # 빈 리스트 선언 방법 1
print('빈 리스트 선언 방법 1',a)
a=[] #빈 리스트 선언 방법 2
print('빈 리스트 선언 방법 2',a)
n=10 # 크기가 n이고, 모든 값이 0인 1차원 리스트 초기화
a = [0] * n
print('크기가 n이고, 모든 값이 0인 1차원 리스트 초기화',a)
```

```python
#리스트의 인덱싱과 슬라이싱
a = [1,2,3,4,5,6,7,8,9] 
print('뒤에서 첫 번째 원소 출력',a[-1]) # 뒤에서 첫 번째 원소 출력
print('뒤에서 세 번째 원소 출력',a[-3]) # 뒤에서 세 번째 원소 출력
a[3] = 7
print('네 번째 원소 값 변경',a)
a = [1,2,3,4,5,6,7,8,9]
print('두 번째 원소부터 네 번째 원소까지',a[1:4]) #두 번째 원소부터 네 번째 원소까지
```

```python
#리스트 컴프리헨션
array = [i for i in range(20) if i % 2 ==1] #0부터 19까지의 수 중에서 홀수만 포함하는 리스트
print('#0부터 19까지의 수 중에서 홀수만 포함하는 리스트',array)
array = [i*i for i in range(1,10)] #1부터 9까지의 수의 제곱 값을 포함하는 리스트
print('1부터 9까지의 수의 제곱 값을 포함하는 리스트',array)
n=3
m=4
array = [[0] * m for _ in range(n)]
print('N X M 크기의 2차원 리스트 초기화',array)
n=3
m=4
array = [[0]*m]*n
print('N X M 크기의 2차원 리스트 초기화'+'('+'잘못된 방법'+')',array)
array[1][1] = 5
print(array)
```

```python
#리스트 관련 메서드
a = [1,4,3]
print('기본 리스트 :',a)
a.append(2) # 리스트에 원소 삽입
print('리스트에 원소 삽입',a)
a.sort() #오름차순 정렬
print('오름차순 정렬 :',a)
a.sort(reverse=True) #내림차순 정렬
print('내림차순 정렬',a)
a.reverse() #리스트 원소 뒤집기
print('원소 뒤집기 : ',a)
a.insert(2,3) # 특정 인덱스에 데이터 추가
print('인덱스 2에 3 추가 :', a)
print('값이 3인 데이터 개수 : ',a.count(3)) # 특정 값인 데이터 개수 세기
a.remove(1) # 특정 값 데이터 삭제
print('값이 1인 데이터 삭제 :',a)
a=[1,2,3,4,5,5,5,] # 특정한 값의 원소 모두 제거
remove_set = {3,5}
result = [i for i in a if i not in remove_set] #remove_set에 포함되지 않은 값 만을 저장
print(result)
```

# 튜플,사전 자료형

```python
### 튜플 자료형
a = (1,2,3,4)
print(a)
# a[2] = 7 ##오류 발생(tuple은 변경 불가능)

### 사전 자료형
data = dict()
data['사과'] = 'Apple'
data['바나나'] = 'Banana'
data['코코넛'] = 'Coconut'
print(data)
if '사과' in data :
    print("'사과'를 키로 가지는 데이터가 존재합니다.")
#사전 자료형 관련 함수
key_list = data.keys() # 키 데이터만 담은 리스트
value_list = data.values() # 값 데이터만 담은 리스트
print('키 데이터만 담은 리스트',key_list)
print('값 데이터만 담은 리스트',value_list)
for key in key_list : # 각 키에 따른 값을 하나씩 출력
    print('각 키에 따른 값을 하나씩 출력',data[key])
```

# 집합 자료형

```python
#집합 자료형 소개
data = set([1,1,2,3,4,4,5]) # 집합 자료형 초기화 방법 1
print(data)
data = {1,1,2,3,4,4,5} # 집합 자료형 초기화 방법 2
print(data)
#집합 자료형의 연산
a = set([1,2,3,4,5])
b = set([3,4,5,6,7])
print('합집합',a | b) #합집합
print('교집합',a & b) #교집합
print('차집합',a - b) #차집합
#집합 자료형 관련 함수
data = set([1,2,3])
print(data)
data.add(4) #새로운 원소 추가
print('새로운 원소 추가',data)
data.update([5,6]) #새로운 원소 여러개 추가
print('새로운 원소 여러개 추가',data) 
data.remove(3) #특정한 값을 갖는 원소 삭제
print('특정한 값을 갖는 원소 삭제',data)
```
