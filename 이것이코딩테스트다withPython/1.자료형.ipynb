# 수 자료형

```python
#정수형
a = 1000 # 양의 정수
print(a)

a = -7 # 음의 정수
print(a)

a = 0 # 0
print(a)
```

```python
#실수형
a = 157.93 # 양의 실수
print('양의 실수', a)
a = -1837.2 # 음의 실수
print('음의 실수',a)
#소수부가 0일 때 0을 생략
a = 5.
print('소수부가 0일 때 0을 생략',a)
#정수부가 0일 때 0을 생략
a = -.7
print('정수부가 0일 때 0을 생략',a)
a = 1e9 # 10억의 지수 표현 방식
print('10억의 지수 표현 방식',a)
a = 75.25e1 # 752.5
print(a)
a= 3954e-3 # 3.954
print(a)

a = 0.3+0.6
print(a)

if a == 0.9:
    print(True)
else :
    print(False)

# round 함수 사용    
a = 0.3 + 0.6
print('a값을 소수점 다섯째 자리에서 반올림',round(a,4))

if round(a,4) == 0.9:
    print(True)
else :
    print(False)
    
# 수 자료형의 연산
a = 7
b = 3
print('나누기',a/b) # 나누기
print('나머지',a%b) # 나머지
print('몫',a // b) # 몫
a = 5
b = 3
print('거듭제곱',a**b) #거듭제곱
```

# 리스트 자료형

```python
#리스트 만들기
a = [1,2,3,4,5,6,7,8,9]
print(a)
print('인덱스 4, 즉 다섯 번째 원소에 접근',a[4]) #인덱스 4, 즉 다섯 번째 원소에 접근
a = list() # 빈 리스트 선언 방법 1
print('빈 리스트 선언 방법 1',a)
a=[] #빈 리스트 선언 방법 2
print('빈 리스트 선언 방법 2',a)
n=10 # 크기가 n이고, 모든 값이 0인 1차원 리스트 초기화
a = [0] * n
print('크기가 n이고, 모든 값이 0인 1차원 리스트 초기화',a)
```

```python
#리스트의 인덱싱과 슬라이싱
a = [1,2,3,4,5,6,7,8,9] 
print('뒤에서 첫 번째 원소 출력',a[-1]) # 뒤에서 첫 번째 원소 출력
print('뒤에서 세 번째 원소 출력',a[-3]) # 뒤에서 세 번째 원소 출력
a[3] = 7
print('네 번째 원소 값 변경',a)
a = [1,2,3,4,5,6,7,8,9]
print('두 번째 원소부터 네 번째 원소까지',a[1:4]) #두 번째 원소부터 네 번째 원소까지
```

```python
#리스트 컴프리헨션
array = [i for i in range(20) if i % 2 ==1] #0부터 19까지의 수 중에서 홀수만 포함하는 리스트
print('#0부터 19까지의 수 중에서 홀수만 포함하는 리스트',array)
array = [i*i for i in range(1,10)] #1부터 9까지의 수의 제곱 값을 포함하는 리스트
print('1부터 9까지의 수의 제곱 값을 포함하는 리스트',array)
n=3
m=4
array = [[0] * m for _ in range(n)]
print('N X M 크기의 2차원 리스트 초기화',array)
n=3
m=4
array = [[0]*m]*n
print('N X M 크기의 2차원 리스트 초기화'+'('+'잘못된 방법'+')',array)
array[1][1] = 5
print(array)
```
